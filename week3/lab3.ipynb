```python
#Problem1
import pandas as pd
import numpy as np

dftuyensinh = pd.read_csv(
    "patient_heart_rate.csv",
    delimiter=",",
    engine="python",
    on_bad_lines="skip"
)
dftuyensinh.head()

column_names = [
    'Id', 'Name', 'Age', 'Weight',
    'm0006', 'm0612', 'm1218',
    'f0006', 'f0612', 'f1218'
]

df = pd.read_csv(
    "patient_heart_rate.csv",
    names=column_names,
    engine="python",
    on_bad_lines="skip"
)

print(df.head())
```

```python
# Problem 2
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
print(df)
```

```python
# Problem 3
# Get the Weight column
weight = df['Weight']

for i in range(0, len(weight)):
    x = str(weight[i])
    
    # In case lbs is part of observation remove it
    if "lbs" in x[-3:]:
        # Remove the lbs from the value
        x = x[:-3]
        
        # Convert string to float
        float_x = float(x)
        
        # Convert to kgs and store as int
        y = int(float_x / 2.2)
        
        # Convert back to string
        y = str(y) + "kgs"
        
        weight[i] = y

print(df)
```

```python
# Problem 4
df.dropna(how="all", inplace=True)
print(df)
```

```python
# Problem 5
df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print(df)
```

```python
# Problem 6
df.Firstname.replace({r'[^\x00-\x7F]+': ''}, regex=True, inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+': ''}, regex=True, inplace=True)
print(df)
```

```python
# Problem 7
# a
print("Số lượng giá trị thiếu:")
print(df[['Age', 'Weight']].isnull().sum())
print("-" * 40)

# b
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# c
df = df.dropna(subset=['Age', 'Weight'], how='all')

# d
mean_age = df['Age'].mean()
print("Giá trị trung bình Age:", round(mean_age, 2))

# e
df['Age'] = df['Age'].fillna(mean_age)

print("\nDữ liệu sau khi xử lý missing values:")
print(df[['Age', 'Weight']])


```

```python
#Problem 8
# Melt the Sex + time range columns in single column
df = pd.melt(
    df,
    id_vars=['Id', 'Age', 'Weight', 'Firstname', 'Lastname'],
    value_name='PulseRate_raw',
    var_name='sex_and_time'
).sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])

# Extract Sex, hour lower bound and hour upper bound
tmp_df = df['sex_and_time'].str.extract(r'(\D)(\d+)(\d{2})', expand=True)

# Name columns
tmp_df.columns = ['Sex', 'hours_lower', 'hours_upper']

# Create Time column
tmp_df['Time'] = tmp_df['hours_lower'] + '-' + tmp_df['hours_upper']

# Merge
df = pd.concat([df, tmp_df], axis=1)

# Drop unnecessary columns
df = df.drop(['sex_and_time', 'hours_lower', 'hours_upper'], axis=1)

# Rename PulseRate column
df.rename(columns={'PulseRate_raw': 'PulseRate'}, inplace=True)

print(df)
```

```python
#11
# Đảm bảo PulseRate là số
df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')

# Khảo sát tỉ lệ dữ liệu thiếu
missing_rate = df['PulseRate'].isna().mean() * 100
print(f"Ti le du lieu thieu PulseRate: {missing_rate:.2f}%")

# Sắp xếp theo từng người và thời gian
df = df.sort_values(by=['Id', 'Time'])

# (1) Trung bình liền trước & liền sau
df['PulseRate'] = df.groupby('Id')['PulseRate'].transform(
    lambda x: x.interpolate(method='linear', limit_direction='both')
)

# (2) Trung bình 2 giá liền trước
df['PulseRate'] = df.groupby('Id')['PulseRate'].transform(
    lambda x: x.fillna(x.shift(1).rolling(2).mean())
)

# (3) Trung bình 2 giá liền sau
df['PulseRate'] = df.groupby('Id')['PulseRate'].transform(
    lambda x: x.fillna(x.shift(-1).rolling(2).mean())
)

# (4) Trung bình huyết áp của người đó
df['PulseRate'] = df.groupby('Id')['PulseRate'].transform(
    lambda x: x.fillna(x.mean())
)

# (5) Trung bình theo nhóm giới tính
df['PulseRate'] = df.groupby('Sex')['PulseRate'].transform(
    lambda x: x.fillna(x.mean())
)

# (6) Trung bình toàn bộ dữ liệu, nếu còn thiếu thì dùng mức ổn định y học
overall_mean = df['PulseRate'].mean()
df['PulseRate'] = df['PulseRate'].fillna(overall_mean)
df['PulseRate'] = df['PulseRate'].fillna(70)
```

```python
#12
# Rút gọn dữ liệu phù hợp
df_clean = df[['Id', 'Age', 'Sex', 'Time', 'PulseRate']]

# Reindex lại dữ liệu
df_clean = df_clean.reset_index(drop=True)

# Lưu file
df_clean.to_csv('patient_heart_rate_clean.csv', index=False)

# Kiểm tra kết quả
print("Kich thuoc du lieu:", df_clean.shape)
print(df_clean.head())

print("Da luu file patient_heart_rate_clean.csv thanh cong")
```

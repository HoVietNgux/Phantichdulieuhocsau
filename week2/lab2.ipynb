```python
#Cau2
dftuyensinh = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

dftuyensinh['phanlopt1'] = np.select(
    [
        dftuyensinh['T1'] < 5,
        (dftuyensinh['T1'] >= 5) & (dftuyensinh['T1'] < 7),
        (dftuyensinh['T1'] >= 7) & (dftuyensinh['T1'] < 8),
        dftuyensinh['T1'] >= 8
    ],
    ['k', 'tb', 'kha', 'g']
)
dftuyensinh[['T1', 'phanlopt1']].head()
```

```python
#Cau3
bang_tanso = dftuyensinh['phanlopt1'].value_counts()
bang_tanso

bang_tanso_df = dftuyensinh['phanlopt1'].value_counts().reset_index()
bang_tanso_df.columns = ['phanlopt1', 'so_luong']

bang_tanso_df
```

```python
#Cau4
import matplotlib.pyplot as plt
grouped = (
    dftuyensinh
    .groupby(['phanlopt1', 'T1'])
    .size()
    .reset_index(name='count')
)

plt.figure(figsize=(10,6))

for label in grouped['phanlopt1'].unique():
    data = grouped[grouped['phanlopt1'] == label]
    plt.plot(data['T1'], data['count'], marker='o', label=label)

plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Multiple Line Chart - T1 theo phân loại phanlopt1")
plt.legend(title="phanlopt1")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Cau5
plt.figure(figsize=(10,6))

for label in grouped['phanlopt1'].unique():
    data = grouped[grouped['phanlopt1'] == label]
    plt.vlines(data['T1'], 0, data['count'], alpha=0.7)
    plt.scatter(data['T1'], data['count'], label=label)

plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Drop-line Chart - T1 theo phân loại phanlopt1")
plt.legend(title="phanlopt1")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#----PHAN5
#Cau1
dftuyensinh['T1'].describe()

plt.figure()
plt.boxplot(dftuyensinh['T1'], vert=False)
plt.title("Boxplot của T1")
plt.xlabel("T1")
plt.show()

dftuyensinh['T1'].skew()

plt.figure()
plt.hist(dftuyensinh['T1'], bins=20)
plt.xlabel("T1")
plt.ylabel("Tần số")
plt.title("Histogram của T1")
plt.show()

dftuyensinh['T1'].kurt()

import scipy.stats as stats

stats.probplot(dftuyensinh['T1'], dist="norm", plot=plt)
plt.title("QQ-Plot của T1")
plt.show()
```

```python
#Câu2
plt.figure(figsize=(8,5))
dftuyensinh.boxplot(column='T1', by='phanlopt1')
plt.title("Boxplot T1 theo phanlopt1")
plt.suptitle("")
plt.xlabel("phanlopt1")
plt.ylabel("T1")
plt.show()

for label in dftuyensinh['phanlopt1'].unique():
    subset = dftuyensinh[dftuyensinh['phanlopt1'] == label]
    plt.hist(subset['T1'], alpha=0.6, bins=10, label=label)

plt.legend()
plt.xlabel("T1")
plt.ylabel("Tần số")
plt.title("Histogram T1 theo phanlopt1")
plt.show()

for label in dftuyensinh['phanlopt1'].unique():
    subset = dftuyensinh[dftuyensinh['phanlopt1'] == label]
    stats.probplot(subset['T1'], dist="norm", plot=plt)
    plt.title(f"QQ-Plot T1 - Nhóm {label}")
    plt.show()

```
